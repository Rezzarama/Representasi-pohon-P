```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Fungsi untuk menambahkan derau garam merica
def add_salt_and_pepper_noise(image, probability=0.05):
    noisy_image = np.copy(image)
    total_pixels = image.size
    num_salt = int(total_pixels * probability / 2)
    num_pepper = int(total_pixels * probability / 2)

    # Menambahkan garam (255) ke citra
    salt_coords = [np.random.randint(0, i, num_salt) for i in image.shape]
    noisy_image[salt_coords[0], salt_coords[1]] = 255

    # Menambahkan merica (0) ke citra
    pepper_coords = [np.random.randint(0, i, num_pepper) for i in image.shape]
    noisy_image[pepper_coords[0], pepper_coords[1]] = 0

    return noisy_image

# Filter Batas (Edge filter)
def boundary_filter(image):
    rows, cols = image.shape
    filtered_image = np.copy(image)

    for i in range(1, rows-1):
        for j in range(1, cols-1):
            neighbors = [
                image[i-1, j-1], image[i-1, j], image[i-1, j+1],
                image[i, j-1], image[i, j], image[i, j+1],
                image[i+1, j-1], image[i+1, j], image[i+1, j+1]
            ]
            min_val = np.min(neighbors)
            max_val = np.max(neighbors)
            if image[i, j] < min_val:
                filtered_image[i, j] = min_val
            elif image[i, j] > max_val:
                filtered_image[i, j] = max_val
            else:
                filtered_image[i, j] = image[i, j]

    return filtered_image

# Filter Pererataan (Averaging filter)
def averaging_filter(image):
    rows, cols = image.shape
    filtered_image = np.copy(image)

    for i in range(1, rows-1):
        for j in range(1, cols-1):
            neighbors = [
                image[i-1, j-1], image[i-1, j], image[i-1, j+1],
                image[i, j-1], image[i, j], image[i, j+1],
                image[i+1, j-1], image[i+1, j], image[i+1, j+1]
            ]
            filtered_image[i, j] = np.mean(neighbors)

    return filtered_image

# Filter Median
def median_filter(image):
    rows, cols = image.shape
    filtered_image = np.copy(image)

    for i in range(1, rows-1):
        for j in range(1, cols-1):
            neighbors = [
                image[i-1, j-1], image[i-1, j], image[i-1, j+1],
                image[i, j-1], image[i, j], image[i, j+1],
                image[i+1, j-1], image[i+1, j], image[i+1, j+1]
            ]
            filtered_image[i, j] = np.median(neighbors)

    return filtered_image

# Load an example image
image_path = '/content/images.jpg'  # Replace with your own image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Tambahkan derau garam merica
noisy_image = add_salt_and_pepper_noise(image, 0.05)

# Terapkan berbagai filter
boundary_filtered_image = boundary_filter(noisy_image)
averaging_filtered_image = averaging_filter(noisy_image)
median_filtered_image = median_filter(noisy_image)

# Menampilkan hasilnya
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes[0, 0].imshow(image, cmap='gray')
axes[0, 0].set_title('Citra Input')
axes[0, 1].imshow(noisy_image, cmap='gray')
axes[0, 1].set_title('Citra Berderau (Garam Merica)')
axes[0, 2].imshow(boundary_filtered_image, cmap='gray')
axes[0, 2].set_title('Filter Batas')

axes[1, 0].imshow(averaging_filtered_image, cmap='gray')
axes[1, 0].set_title('Filter Pererataan')
axes[1, 1].imshow(median_filtered_image, cmap='gray')
axes[1, 1].set_title('Filter Median')

plt.show()
```
